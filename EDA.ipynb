# EDA

Exploratory data analysis (EDA) is used by data scientists to analyze
and investigate data sets and summarize their main characteristics, 
often employing data visualization methods.

## Univariate,Bivariate and MultiVariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiu+c-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df.shape
```

### Univariate Analysis

Univariate statistics summarize only one variable at a time.


```python
df_setosa=df.loc[df['species']=='setosa']
```

```python
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Petal length')
plt.show()
```

### Bivariate Analysis

Bivariate statistics compare two variables.

```python
sns.FacetGrid(df,hue="species",size=5).map(plt.scatter,"petal_length","sepal_width")
plt.legend()
plt.show()
```

### Multivariate Analysis

Multivariate statistics compare more than two variables.

```python
sns.pairplot(df,hue="species",size=3)
```
